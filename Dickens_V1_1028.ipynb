```python
# -*- coding: utf-8 -*-
"""

合并一个文件夹下的多个txt文件

"""
import os
time1=time.time()
os.chdir(r"C:/Users/XPS/Desktop/第一批训练文本")
# 获取目标文件夹的路径
filedir = "C:/Users/XPS/Desktop/第一批训练文本"
# 获取当前目录下的文件名列表
filename = os.listdir(filedir)
# 打开当前目录下的 result1.txt 文件 如果不存在则创建
f = open('result1.txt', 'w',encoding='UTF-8')
i = 0
# 先遍历文件名
for filename in filename:
    i += 1
    print(i)
    if i > 0:
        filepath = filedir + '/' + filename
        print(filepath)
        # 遍历单个文件 读取行数
        for line in open(filepath, encoding='UTF-8'):
            # print(str(line))
            f.writelines(line)
            # f.write('\n')
# 关闭文件
f.close()
time2 = time.time()
print(u'总共耗时：' + str(time2 - time1) + 's')
```

```python
from nltk import word_tokenize     #以空格形式实现分词
fffwords = word_tokenize(fffstring)
a_words = fffwords[:1000]
print(a_words)
```

```python
from nltk import word_tokenize     #以空格形式实现分词
eeewords = word_tokenize(eeestring)
```

```python
from nltk import sent_tokenize    #以符号形式实现分句, 本次不选用！！！
fffsentences = sent_tokenize(fffstring)
a_fffsentences = fffsentences[:10000]
print(a_fffsentences)
```

```python
interpunctuations = [',', '.', ':', ';', '?', '(', ')', '[', ']', '&', '!', '*', '@', '#', '$', '%']   #定义标点符号列表
```

```python
fffwordscut = [word for word in fffwords if word not in interpunctuations]   #去除标点符号
print('\n【NLTK分词后去除符号结果：】')
a_wordscut = fffwordscut[:1000]
print(a_wordscut)
```

```python
eeewordscut = [word for word in eeewords if word not in interpunctuations]   #去除标点符号
```

```python
from nltk.corpus import stopwords
```

```python
stops = set(stopwords.words("english"))
fffwordscut2 = [word for word in fffwordscut if word not in stops]
print('\n【NLTK分词后去除停用词结果：】')
a_wordscut2 = fffwordscut2[:1000]
print(a_wordscut2)
```

```python
stops = set(stopwords.words("english"))
eeewordscut2 = [word for word in eeewordscut if word not in stops]
```

```python
from nltk import pos_tag
```

```python
fffwordscut3 = pos_tag(fffwordscut2)
print('\n【NLTK分词后词性标注结果：】')
a_wordscut3 = fffwordscut3[:1000]
print(a_wordscut3)
```

```python
eeewordscut3 = pos_tag(eeewordscut2)
```

```python
##列表中每个元组第二个元素显示为该词的词性
##nltk.help.upenn_tagset()可以用来查询词性标签含义
##或者参看文档https://blog.csdn.net/sk_berry/article/details/105267685
```

```python
from nltk.stem import PorterStemmer    #词干提取
from nltk.stem import WordNetLemmatizer    #词性还原
```

```python
fffwordscut4 = []
for fffwordscut in fffwordscut2:
    fffwordscut4.append(PorterStemmer().stem(fffwordscut))    #词干提取
a_wordscut4 = fffwordscut4[:10000]
print(a_wordscut4)
```

```python
eeewordscut4 = []
for eeewordscut in eeewordscut2:
    eeewordscut4.append(PorterStemmer().stem(eeewordscut))    #词干提取
eeewordscut5 = []
for eeewordscut in eeewordscut4:
    eeewordscut5.append(WordNetLemmatizer().lemmatize(eeewordscut,pos='v'))   #指定还原词性为名词
```

```python
print('\n【NLTK分词进行词形还原：】')
fffwordscut5 = []
for fffwordscut in fffwordscut4:
    fffwordscut5.append(WordNetLemmatizer().lemmatize(fffwordscut,pos='v'))   #指定还原词性为名词
a_wordscut5 = fffwordscut5[:1000]
print(a_wordscut5)
```
